```python
import pandas as pd
df=pd.read_csv('PostComputationData.csv')
df.head()
```

```python
year_list=[1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
dw=[]
bw=[]
tb=[]


for year in year_list:
    df_new=df[df['YEAR_OBS']==year]
    DF=df_new[df_new['SITE_TYPE']=='Dug Well']
    dw.append(DF.POMRB.mean())


for year in year_list:
    df_new=df[df['YEAR_OBS']==year]
    DF=df_new[df_new['SITE_TYPE']=='Bore Well']
    bw.append(DF.POMRB.mean())
    

for year in year_list:
    df_new=df[df['YEAR_OBS']==year]
    DF=df_new[df_new['SITE_TYPE']=='Tube Well']
    tb.append(DF.POMRB.mean())
    
```

```python
for x in range(len(dw)):
    print(dw[x])
```

```python
for x in range(len(bw)):
    print(bw[x])
```

```python
for x in range(len(tb)):
    print(tb[x])
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
#n_groups = 21


 
# create plot
#fig, ax = plt.subplots()
index = np.arange(len(year_list))
bar_width = 0.45
opacity = 0.8
 
plt.bar(index+0.0, dw, width=0.25,
                 alpha=opacity,
                 color='r',
                 label='Dug Well')
plt.bar(index+0.25 , bw, width=0.25,
                 alpha=opacity,
                 color='b',
                 label='Bore Well')
plt.bar(index+0.50, tb, width=0.25,
                 alpha=opacity,
                 color='g',
                 label='Tube Well')
 
plt.xlabel('YEARS')
plt.ylabel('Values')
plt.title('POST MONSOON WATER LEVEL BY WELL SITE')
plt.xticks(index + 0.25, (year_list))
plt.legend()
plt.tight_layout()
plt.rc('xtick', labelsize=3) 
plt.savefig("Post Monsoon Rabi.png",dpi=200,quality=95,bbox_inches='tight')
#plt.figure(figsize=(440,120))
plt.show()
```
