# filter the rows where the YEAR_OBS is 1996

```python
f2=f1[f1['YEAR_OBS']==1996]
f2
```

# calculate the mean of the 'MONSOON' attribute of all the rows where the YEAR_OBS is 1996

```python
m=f2['MONSOON'].mean()
m
```

# calculate the mean of the attribute 'MONSOON' for all the states 

```python
state_list=['AP','AR','AS','BR','CG','GJ','HR','JK','JH','KA','KL','MP','ML','OD','PB','RJ','TN','TR','CH','DL']
mean_states=[]
for state in state_list:
    f_new=f2[f2['STATE']==state]
    mean_states.append(f_new.MONSOON.mean())
```

```python
for x in range(len(mean_states)):
    print(mean_states[x])
```

```python
most=[]
least=[]
for mean in mean_states:
    if mean>=m:
        most.append(mean)
    else:
        least.append(mean)
```

```python
for x in range(len(most)):
    print(most[x])
```

```python
for x in range(len(least)):
    print(least[x])
```
