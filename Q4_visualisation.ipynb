```python
import pandas as pd            #Load dataset
df=pd.read_csv('PostComputationData.csv')
df.head()
```

```python
#calculating the average of rabi effect and kharif effect over years
year_list=[1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
rb=[]
kh=[]
for year in year_list:
    df_new=df[df['YEAR_OBS']==year]
    rb.append(df_new.RB_Effect.mean())
    kh.append(df_new.KH_Effect.mean())
```

```python
#average of rabi effect
for x in range(len(rb)):
    print(rb[x])
```

```python
#average of kharif effect
for x in range(len(kh)):
    print(kh[x])
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import *

years=[1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
plt.plot(years, rb , color='red',label='Effect of Groundwater Withdrawal for Rabi Crops')
plt.plot(years, kh , color='blue',label='Effect of Groundwater Withdrawal for Kharif Crops')
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Years')
plt.ylabel('Average GroundWater Level')
plt.title('Effect of Groundwater Withdrawal For Rabi and Kharif Crops')
plt.xticks(years)
plt.rc('xtick', labelsize=10) 
plt.savefig("Figures/GroundwaterEffect.png",dpi=200,quality=95,bbox_inches='tight')
plt.show()
```
